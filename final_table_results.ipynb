```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import ipdb
```

```python
def get_table_of_det(det):
    models = ['RF', 'SVR', 'XGB', 'NN']
    table = pd.DataFrame(
        [], columns=[['RF base', 'RF bda', 'SVR base', 'SVR bda', 'XGB base', 'XGB bda', 'NN base', 'NN bda']], 
        index=[['NRMSE', 'SMAPE', 'NMAE']]
    )

    for model in models:
        base = pd.read_csv('./table_data/base_'+model+'_mean.csv')
        bda = pd.read_csv('./table_data/bda_'+model+'_mean.csv')

        table.loc['NRMSE', model+' base'] = base.loc[det, 'nrmse']
        table.loc['NRMSE', model+' bda'] = bda.loc[det, 'nrmse']

        table.loc['SMAPE', model+' base'] = base.loc[det, 'smape']
        table.loc['SMAPE', model+' bda'] = bda.loc[det, 'smape']

        table.loc['NMAE', model+' base'] = base.loc[det, 'nmae']
        table.loc['NMAE', model+' bda'] = bda.loc[det, 'nmae']

    return table
```
